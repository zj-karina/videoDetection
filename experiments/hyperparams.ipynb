```python
from ultralytics import YOLO

model = YOLO("yolov5x.pt")
```

```python
search = model.tune(data='/content/drive/MyDrive/datasets/dataset.yaml',
                         epochs=100,
                        use_ray = True,
                         batch=8,
                         workers=16,
                         imgsz = 415,
                         optimizer='Adam',
                         seed=0,
                         pretrained=True,
                         space={"lr0": tune.uniform(1e-5, 1e-1),
                               "momentum": tune.uniform(0.6, 0.98),
                               "weight_decay": tune.uniform(0.0, 0.001)
                               },
                         )
```
